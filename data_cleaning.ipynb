```python
# Importing libraries

import pandas as pd
import numpy as np
import mysql.connector
import getpass
```

```python
# Database password request

password = getpass.getpass()
```

```python
# I establish the connection with the database and check if it was successful

cnx = mysql.connector.connect(user = 'root', password = password, database = 'ih_gambling', port=3306)
cnx.is_connected()
```

```python
# Definition of the cursor object to interact with the database

cursor = cnx.cursor()
```

# Account table

```python
# I instantiate the query to get the data from the table, run it, and finally save the data in the results variable.

query = ("SELECT * FROM ih_gambling.account;")
cursor.execute(query)
results = cursor.fetchall()
```

```python
# I create the pandas dataframe with the data so I can clean and analyze it.

account_df = pd.DataFrame(results)
account_df
```

```python
# Renaming column names

account_df.columns = ['AccountNo', 'CustId', 'AccountLocation', 'CurrencyCode', 'DailyDepositLimit', 'StakeScale', 'SourceProd']
account_df
```

```python
# Check whether null values are present

account_df.isnull().sum()
```

```python
account_df.info()
```

```python
# Creation of the csv file of the table in order to report it to sql.

account_df.to_csv('csv_file/account.csv', index=False)
```

# Betting table

```python
query = ("SELECT * FROM ih_gambling.betting;")
cursor.execute(query)
results = cursor.fetchall()
```

```python
betting_df = pd.DataFrame(results)
betting_df.head()
```

```python
betting_df.columns = ['AccountNo', 'BetDate', 'ClassId', 'CategoryId', 'Source', 'BetCount', 'Bet_Amt', 'Win_Amt', 'Product', 'MyUnknownColumn', 'MyUnknownColumn_[0]', 'MyUnknownColumn_[1]', 'AccountNo_[0]', 'Bet_Amt_[0]', 'Product_[0]', 'MyUnknownColumn_[2]', 'AccountNo_[1]', 'Vegas', 'Sportsbook', 'Games', 'Casino', 'Poker', 'Bingo', '#N/A', 'Adjustments']
betting_df.head()
```

```python
# Removal of duplicates

betting_df.drop_duplicates(inplace=True)
betting_df
```

```python
# Remove unnecessary columns (if not relevant to the analysis).

betting_df.drop(columns=['MyUnknownColumn', 'MyUnknownColumn_[0]', 'MyUnknownColumn_[1]', 'MyUnknownColumn_[2]', 'AccountNo_[0]', 'AccountNo_[1]', 'Bet_Amt_[0]', 'Product_[0]'], inplace=True)
```

```python
betting_df.head()
```

```python
betting_df.isnull().sum()
```

```python
# Column conversion to datetime

betting_df['BetDate'] = pd.to_datetime(betting_df['BetDate'], errors='coerce', dayfirst=True)
```

```python
betting_df.info()
```

```python
# Unique values in ClassId and CategoryId
print(betting_df['ClassId'].unique())
print(betting_df['CategoryId'].unique())
```

```python
# Replace empty values with NaN

betting_df['ClassId'] = betting_df['ClassId'].replace('', pd.NA)
```

```python
betting_df.isnull().sum()
```

```python
# Remove rows with null values in ClassId

betting_df = betting_df.dropna(subset=['ClassId'])
```

```python
# Replaces ' - ' with NaN in all columns of the dataframe

betting_df.replace(' -   ', np.nan, inplace=True)
betting_df.head()
```

```python
betting_df.to_csv('csv_file/betting.csv', index=False)
```
